```python
# archive here https://archive.ph/MhzS1
url1 = "https://raw.githubusercontent.com/ActiveConclusion/COVID19_mobility/master/google_reports/mobility_report_countries.csv"
google = pd.read_csv(url1)
```

```python
# Let's inspect our data a bit
google.head()
google.sample(20)
```

```python
# Let's now restrict our sample to only the countries of interest and drop subregions 
# since not all data-sets have those enumerated
google = google.loc[google['country'].isin(countries)]
google = google.loc[google["region"] == "Total"]
google.head()
google.sample(20)
google.describe()
```

```python
# now we do the same for the Apple data
# archive here: archive.ph/4ljCg
url2 = "https://raw.githubusercontent.com/ActiveConclusion/COVID19_mobility/master/apple_reports/apple_mobility_report.csv"
apple = pd.read_csv(url2) 
apple = apple.loc[apple["country"].isin(countries)]
```

```python
apple.sample(10)
# It does look like there are some sub-regions here, so let's drop
apple = apple.loc[apple["sub-region"] == "Total"]
apple.sample(10)
apple.groupby("country").mean()
```

```python
# Time to pull the mortality data
url3 = "https://raw.githubusercontent.com/akarlinsky/world_mortality/main/world_mortality.csv"
mortality = pd.read_csv(url3)
mortality.head()
```

```python
# And, let's do the same thing. I'm also going to rename columns now. 
mortality = mortality.loc[mortality["country_name"].isin(countries)]
mortality["country"] = mortality["country_name"]
mortlality = mortality.drop(columns=["country_name"])
mortality["week"] = mortality["time"]
mortality = mortality.drop(columns=["time", "time_unit"])
mortality.head()
```

```python
# First, let's merge the daily data and convert to time-series format while pulling 
# information on the year and the week-of-year
df = pd.merge(apple, google, how="outer", on=["country", "date"])
df["date"] = pd.to_datetime(df["date"])
df["week"] = df["date"].dt.isocalendar().week
df["year"] = df["date"].dt.isocalendar().year

# Now we can merge on the unique pairings of country, year, and week 
df = pd.merge(df, mortality, how="outer", on=["country", "week", "year"])

# And drop some unnecessary variables
df = df.drop(columns=["sub-region", "subregion_and_city", "geo_type", "iso3c", "country_name"])

# And reorder some key columns while leaving the others where they are
keycols = ["country", "date", "year", "week"]
cols = keycols  + [col for col in df if col not in keycols]
df = df[cols]
df.head(25)
```

```python
# Let's now make sure that our data-set is sorted and take a peak with a sorted sample. 
# Note that we have to re-set the index manually in Pandas.
df = df.sort_values(by =["country", "year", "date"])
df = df.reset_index(drop=True)
df.head(50)
# This is good -- we should only expect excess mortality data from those earlier years
```

```python
# The next step is to make some weekly excess death figures.
# Note that the object that this actually creates is basically a Kronecker decomposition
# We get a list of country-week pairs with deaths for each observation.
# This is explained well here: https://archive.ph/XDSlJ
# Note that the first few observations for, e.g., Australia, Week 1 have distinct values 
# because the daily data-sets don't go that far back
# But then later on, there are repeats, reflecting the start of daily data. 
pre20data = df[df["year"]<2020]
gb = pre20data.groupby(["country", "week"])["deaths"]
for i in gb:
    print(i)
```

```python
# Now, we want to take those means 
# This is quite helpful. https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_stata.html
df["week_mean_death"] = gb.transform("mean")
```

```python
df["week_mean_death"] = df.groupby("country")["week_mean_death"].ffill()
df["weekly_excess_death"] = df["deaths"] - df["week_mean_death"]
df["excess_mortality"] = (df["weekly_excess_death"]*100)/df["week_mean_death"]
df.head(50)
df.sample(50)
```

```python
keyvars = "date", "driving", "transit", "walking", "excess_mortality"
df_mod = df[df["year"]>2019]
df_mod = df_mod[["date", "country", "driving", "transit", "walking", "excess_mortality" ]]
df_melt = pd.melt(df_mod, ["country", 'date'])
df_melt.head(20)
```
