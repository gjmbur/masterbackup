```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.tan(x))
ax[1].plot(x, np.sqrt(x))
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T # VanderPlas doesn't really explain this but this is the transpose of the data matrix. 
# I'm not completely sure why that's necessary here, though. 
```

```python
# Here's how we can print the variable names
print(iris['feature_names']) # see excellent answer here https://archive.ph/ZkPfS
```

```python
# Here's how we can print the data matrix
print(iris['data'])
```

```python
# ... and now one column of its transpose
print(features[0])
```

```python
# Here's the full data-set description
print(iris['DESCR'])
```

```python
# plt.scatter's syntax is plt.scatter(xvar, yvar)
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
