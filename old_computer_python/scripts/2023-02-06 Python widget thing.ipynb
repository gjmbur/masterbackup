```python
# Replace the expression in quotation marks with your username
user = "GJMB"
baseurl = "https://wwwn.cdc.gov/NCHS/NHANES/2017-2018/"
folder = f"/Users/{user}/desktop/Python/data/NHANES/"
if not(os.path.exists(folder) and os.path.isdir(folder)):
    os.makedirs(folder)
```

```python
#Import the first file before starting a loop
firstfile = folder + "DEMO_J.XPT"
firsturl = baseurl + "DEMO_J.XPT"
data = pd.read_sas(firsturl, format="xport")
```

```python
stubs = ["BMX_J", "BPX_J", "CDQ_J", "PAQ_J", "SMQ_J", "SMQFAM_J", "MCQ_J", "SMQRTU_J", "SMQSHS_J", "UM_J"]
for i in stubs:
    url = baseurl + i + ".XPT"
    data_temp = pd.read_sas(url, format="xport")
    data = pd.merge(data, data_temp)
```

```python
for cols in data.columns:
    print(cols)
```

```python
merged = folder + "merged_NHANES"
pyreadstat.write_xport(data, merged)
```
